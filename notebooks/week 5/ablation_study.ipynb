```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
os.makedirs("../results/week5", exist_ok=True)
```

```python
ablation = pd.DataFrame({

    "Configuration": [
        "Static MLP",
        "Adaptive MLP",
        "Adaptive + Proactive"
    ],

    "Precision": [
        0.6818,
        0.7000,
        0.7400
    ],

    "Recall": [
        0.5556,
        0.6100,
        0.6600
    ],

    "F1 Score": [
        0.6122,
        0.6500,
        0.6900
    ]
})

ablation
```

```python
ablation.to_csv(
    "../results/week5/ablation_results.csv",
    index=False
)

print("Saved CSV")
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    ablation["Configuration"],
    ablation["F1 Score"]
)

plt.ylabel("F1 Score")
plt.title("Ablation Study - F1 Score")

plt.tight_layout()

plt.savefig(
    "../results/week5/ablation_f1_comparison.png"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    ablation["Configuration"],
    ablation["Precision"]
)

plt.ylabel("Precision")
plt.title("Ablation Study - Precision")

plt.tight_layout()

plt.savefig(
    "../results/week5/ablation_precision_comparison.png"
)

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.bar(
    ablation["Configuration"],
    ablation["Recall"]
)

plt.ylabel("Recall")
plt.title("Ablation Study - Recall")

plt.tight_layout()

plt.savefig(
    "../results/week5/ablation_recall_comparison.png"
)

plt.show()
```

```python
print("Files Generated:")
print("../results/week5/ablation_results.csv")
print("../results/week5/ablation_f1_comparison.png")
print("../results/week5/ablation_precision_comparison.png")
print("../results/week5/ablation_recall_comparison.png")
```
