```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../data/diabetes.csv")
df.head() 
```

```python
features = df.columns[:-1] 
df[features].hist(figsize=(15,10))
plt.tight_layout()
plt.show()
```

```python
class_counts = df["Outcome"].value_counts()
print(class_counts)
```

```python
class_counts.plot(kind="bar")
plt.title("Pima Diabetes Class Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks([0,1], ["Non-Diabetic", "Diabetic"], rotation=0)

plt.savefig(
    "../../results/week 1/class_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
